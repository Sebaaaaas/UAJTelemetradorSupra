```python
print(x)
print(y)
```

```python
w,b,J = gradient_descent(x,y,0,0,compute_cost,compute_gradient,0.01,1500)
print(w)
print(b)
print(J)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(J)), J, marker='o', linestyle='-', color='b')
plt.xlabel("Iteraciones")
plt.ylabel("J (coste)")
plt.show()
```

```python
import numpy as np
w_final=w[w.size-1]
x_regression = np.array([0,25])
y_regression = w_final * x_regression + b

# Mostrar los datos
plt.scatter(x, y, label="Datos")

# Recta de regresion

regression_line = w_final * x_regression + b
plt.plot(x_regression, regression_line, color='red', label="Recta de Regresión")

plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()

plt.show()
```
